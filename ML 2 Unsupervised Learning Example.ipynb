# Preprocessing

```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### 10.a. Generating data set with 50 variables and 60 observations (with 20 observations in each of three classes)

```python
#Generate data frame of random numbers
X = pd.DataFrame()
const = np.random.randint(low=100, size = 1)

for num in range(0,50):
    Z = np.random.normal(loc=0.0, scale=0.01, size = 60)
    X = X.append(pd.DataFrame([Z]))

df = X.T
cols = np.linspace(0,49, num = 50).astype(int)
df.columns = cols

df.head()
```

```python
#splitting up observations with 20 in one of three classes
df[1].iloc[0:20] = 1
df[0].iloc[20:40] = 2
df[1].iloc[40:60] = 3
df.head()
```

### 10.b. Performing PCA on the 60 observations

```python
import warnings
warnings.filterwarnings('ignore')

from sklearn.preprocessing import scale
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=50)
pca_data = pca.fit_transform(df)
principaldf = pd.DataFrame(data = pca_data)
principaldf.head()
```

```python
loadings = pca.components_.T
loadings_df = pd.DataFrame(loadings, index=df.columns, columns=principaldf.columns)
loadings_df.head()
```

```python
plt.figure(figsize=(25, 10))
plt.scatter(principaldf[0], principaldf[1], s=500)
plt.title('principal components', fontsize=30, color='k')
plt.xlabel('principal component 1', fontsize=20, color='g')
plt.ylabel('principal component 2', fontsize=20, color='g')
for length, vector in zip(pca.explained_variance_, pca.components_):
    v = vector * 5e-17* np.sqrt(length)
    arrow = plt.arrow(0, 0, pca.mean_[0] + v[0], pca.mean_[1] + v[1], 
                      label='mylabel', width=0.005, 
                      facecolor='orange', 
                      edgecolor='orange', 
                      alpha=0.5, )
```

### 10.c K-means clustering of the observations with K = 3

```python
from sklearn.cluster import KMeans as KM
from sklearn.metrics import confusion_matrix, classification_report
```

```python
km_out = KM(n_clusters = 3, n_init=20).fit(df)
km_pred = km_out.predict(df)
```

```python
km_labels = pd.DataFrame(km_out.labels_)
km_labels[0].value_counts()
```
